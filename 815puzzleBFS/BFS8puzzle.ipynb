```python
from collections import deque

class PuzzleState:
    def __init__(self, puzzle):
        self.puzzle = puzzle
        self.size = len(puzzle)
        self.blank = self.find_blank()

    def __str__(self):
        return '\n'.join([' '.join(map(str, row)) for row in self.puzzle])

    def find_blank(self):
        for i in range(self.size):
            for j in range(self.size):
                if self.puzzle[i][j] == 0:
                    return i, j

    def move(self, direction):
        i, j = self.blank
        if direction == 'up' and i > 0:
            self.puzzle[i][j], self.puzzle[i - 1][j] = self.puzzle[i - 1][j], self.puzzle[i][j]
            self.blank = (i - 1, j)
        elif direction == 'down' and i < self.size - 1:
            self.puzzle[i][j], self.puzzle[i + 1][j] = self.puzzle[i + 1][j], self.puzzle[i][j]
            self.blank = (i + 1, j)
        elif direction == 'left' and j > 0:
            self.puzzle[i][j], self.puzzle[i][j - 1] = self.puzzle[i][j - 1], self.puzzle[i][j]
            self.blank = (i, j - 1)
        elif direction == 'right' and j < self.size - 1:
            self.puzzle[i][j], self.puzzle[i][j + 1] = self.puzzle[i][j + 1], self.puzzle[i][j]
            self.blank = (i, j + 1)
        else:
            return False  # Invalid move
        return True

def bfs(initial_state, final_state):
    queue = deque([(initial_state, [])])
    visited = set()
    count_state = 0
    while queue:
        current_state, path = queue.popleft()

        if current_state.puzzle == final_state.puzzle:
            print(f"Depth {len(path)}:")
            count_state += 1
            print(current_state)
            return path, count_state

        if tuple(map(tuple, current_state.puzzle)) not in visited:
            visited.add(tuple(map(tuple, current_state.puzzle)))

            for direction in ['up', 'down', 'left', 'right']:
                new_state = current_state.__class__([row[:] for row in current_state.puzzle])
                if new_state.move(direction):
                    if tuple(map(tuple, new_state.puzzle)) not in visited and (new_state, path + [direction]) not in queue:
                        if new_state.puzzle == final_state.puzzle:
                            print(f"Depth {len(path) + 1}:")
                            print(new_state)
                            count_state += 1
                            return path + [direction], count_state
                        queue.append((new_state, path + [direction]))
                        count_state += 1

                        # Print nodes at every depth
                        print(f"Depth {len(path) + 1}:")
                        print(new_state)
    return None, count_state  # No solution found






def getInvCount(arr):
    inv_count = 0
    empty_value = -1
    for i in range(0, 9):
        for j in range(i + 1, 9):
            if arr[j] != empty_value and arr[i] != empty_value and arr[i] > arr[j]:
                inv_count += 1
    return inv_count


def isSolvable(puzzle) :
    inv_count = getInvCount([j for sub in puzzle for j in sub])
    return (inv_count % 2 == 0)

def main():
    initial_state = PuzzleState([[2, 0, 4], [1, 5, 6], [7, 8, 3]])
    puzzle=[[2, 0, 4], [1, 5, 6], [7, 8, 3]]
    if(isSolvable(puzzle)) :
      print("Solvable")
    else :
      print("Not Solvable due to odd inversions")
      return
    final_state = PuzzleState([[2, 4, 6], [1, 0, 5], [7, 8, 3]])

    print("Initial State:")
    print(initial_state)

    print("Goal State:")
    print(final_state)

    solution,state = bfs(initial_state, final_state)
    if solution:
        print("\nSolution:")
        for step, direction in enumerate(solution, start=1):
            initial_state.move(direction)
            print(f"Depth {step}: Move {direction}")
            print(initial_state)
        print("Total States:",state)

    else:
        print("\nNo solution found")

if __name__ == "__main__":
    main()
```
