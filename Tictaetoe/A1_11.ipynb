```python
matrix = []
count = 0
import random

for i in range(3):
    row = []
    for j in range(3):
        row.append(0)
    matrix.append(row)
print(matrix)

def computer():
    print("Computer Playing")
    while True:
        row = random.randint(0, 2)
        col = random.randint(0, 2)
        if matrix[row][col] == 0:
            matrix[row][col] = 'o'
            return

def user():
    print("User Playing")
    while True:
        try:
            row = int(input("Enter row (0, 1, or 2): "))
            col = int(input("Enter col (0, 1, or 2): "))
            if 0 <= row <= 2 and 0 <= col <= 2 and matrix[row][col] == 0:
                matrix[row][col] = 'x'
                return
            else:
                print("Invalid move. Try again.")
        except ValueError:
            print("Invalid input. Please enter a number.")

def check_winner():
    # row validation
    for i in range(3):
        if matrix[i][0] == matrix[i][1] == matrix[i][2] and matrix[i][0] != 0:
            return matrix[i][0]

    # diagonal validation
    if matrix[0][0] == matrix[1][1] == matrix[2][2] and matrix[0][0] != 0:
        return matrix[0][0]
    if matrix[0][2] == matrix[1][1] == matrix[2][0] and matrix[0][2] != 0:
        return matrix[0][2]

    # column validation
    for i in range(3):
        if matrix[0][i] == matrix[1][i] == matrix[2][i] and matrix[0][i] != 0:
            return matrix[0][i]

    return None

while count <= 8:
    user()
    print("Final matrix:\n")
    for i in range(3):
        for j in range(3):
            print(matrix[i][j], end=' ')
        print("\n")

    winner = check_winner()
    if winner == 'x':
        print("Player 1 wins\n")
        break
    elif winner == 'o':
        print("Computer wins\n")
        break

    if count == 8:
        print("\nGAME DRAWN\n")
        break

    computer()
    print("Final matrix:\n")
    for i in range(3):
        for j in range(3):
            print(matrix[i][j], end=' ')
        print("\n")
    winner = check_winner()
    if winner == 'x':
        print("Player 1 wins\n")
        break
    elif winner == 'o':
        print("Computer wins\n")
        break

    if count == 8:
        print("\nGAME DRAWN\n")
        break

    count += 2
```
