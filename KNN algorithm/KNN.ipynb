```python
import math

def classifyAPoint(points, p, k=3):
    distance = []
    for group in points:
        for feature in points[group]:
            # Calculate the Euclidean distance of p from training points
            euclidean_distance = math.sqrt((feature[0] - p[0]) ** 2 + (feature[1] - p[1]) ** 2)
            # Add a tuple of form (distance, group) in the distance list
            distance.append((euclidean_distance, group))
    # Sort the distance list in ascending order based on distance
    distance = sorted(distance, key=lambda x: x[0])
    print(distance)
    # Select first k distances
    distance = distance[:k]
    print(distance)
    # Count frequencies of each group
    freq = {0: 0, 1: 0}
    for d in distance:
        freq[d[1]] += 1
    return 0 if freq[0] > freq[1] else 1

# Driver function
def main():
    # Dictionary of training points having two keys - 0 and 1
    # Key 0 has points belong to class 0
    # Key 1 has points belong to class 1
    points = {
        0: [(1, 12), (2, 5), (3, 6), (3, 10), (3.5, 8), (2, 11), (2, 9), (1, 7)],
        1: [(5, 3), (3, 2), (1.5, 9), (7, 2), (6, 1), (3.8, 1), (5.6, 4), (4, 2), (2, 5)]
    }
    # Testing point p(x,y)
    p = (2.5, 7)
    # Number of neighbours
    k = 3
    print("The value classified to unknown point is: {}".format(classifyAPoint(points, p, k)))

if __name__ == '__main__':
    main()
```
